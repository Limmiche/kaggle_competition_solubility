# Descriptor histograms
to check for features that could allow for a better prediction, histograms of all descriptors were plotted to see wheter some if any show a separation of the three classes

```python
import pandas as pd
import numpy as np
from matplotlib import pyplot as plot
```

```python
df = pd.read_csv("output/df_train_set.csv")
```

```python
df.drop(columns=['Id', 'smiles'], inplace=True)
df.drop([column for column in df.columns if column.startswith('ecfp_')], axis=1, inplace=True)
```

the three solubility classes are separated into individual data sets

```python
high = df[df['sol_category'] == 2].copy(deep=True)
medium = df[df['sol_category'] == 1].copy(deep=True)
low = df[df['sol_category'] == 0].copy(deep=True)
```

```python
df.drop(columns=['sol_category'], inplace=True)
high.drop(columns=['sol_category'], inplace=True)
medium.drop(columns=['sol_category'], inplace=True)
low.drop(columns=['sol_category'], inplace=True)
```

```python
def plotHistogram(x,y,z):
    plot.hist([x, y, z], color=['g','r', 'b'], alpha=0.5, bins=12)
    plot.show()
```

```python

#print('--------------------------------------------------------------------------')

for col in df.columns:
    #value_range = df[col]
    high_solubility = high[col]
    medium_solubility = medium[col]
    low_solubility = low[col]

    print('-------------------------------------------------------------------------------')
    print(f'histogram of descriptor {col}')

    plotHistogram(high_solubility,medium_solubility,low_solubility)

    
```
