## Removing unused bits

```python
for (colname,colval) in df_train.items():
    if colname.startswith("ecfp_"):
        value = 0
        for colval in df_train[colname]:
            value += colval
        #print(value)
        if value == 0:
            #print('delete column', colname)
            df_train.drop(columns=colname, inplace=True)
```
