## Removing unused columns

```python
# to do: change for loop to pandas commands
```

```python
len_un = len(df.columns)
df = df.loc[:,df.nunique()!=1]
num_drop_unique = len_un - len(df.columns)
print(f'number of dropped collumns with identical values: {num_drop_unique}')
```

## Correlation

```python
# Korrelierte Größen filtern
```

```python
df_corr = df.copy(deep=True)
df_corr.drop(columns='sol_category', inplace=True)
df_corr
```

#### Removing ecfp from copy

```python
drop_list = [column for column in df_corr.columns if column.startswith('ecfp_')]

# Drop features 
df_corr.drop(drop_list, axis=1, inplace=True)
df_corr
```

#### Creating correlation matrix

```python
# Create correlation matrix
corr_matrix = df_corr.corr(numeric_only=True).abs()

# Select upper triangle of correlation matrix
upper = corr_matrix.where(np.triu(np.ones(corr_matrix.shape), k=1).astype(bool))
```

#### Dropping highy correlated columns

```python
# Find features with correlation greater than 0.95
to_drop = [column for column in upper.columns if any(upper[column] > 0.95)]

# Drop features 
df.drop(to_drop, axis=1, inplace=True)
```

```python
len_final = len(df.columns)
len_dropped = len_start - len_final
num_desc_final = sum([1 for col_name in df.columns if 'example_' in col_name  or 'rdkit_desc_' in col_name])
num_desc_final

print(f'number of columns at the beginning: {len_start}')
print(f'number of columns at the end: {len_final}')
print('------------------------------------------------------')
print(f'total number of descriptors at the beginning: {num_desc_total}')
print(f'total number of descriptors at the end: {num_desc_final}')
print(f'total number of dropped descriptors: {len_dropped}')
print('------------------------------------------------------')
print(f'number of dropped collumns with identical values: {num_drop_unique}')
print(f'number of dropped descriptors due to high correlation > 0.95: {len(to_drop)}')
print('------------------------------------------------------')
print('dropped descriptors:')
print(to_drop)
```
